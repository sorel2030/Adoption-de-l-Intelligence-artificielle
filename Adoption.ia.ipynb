```python
from google.colab import drive
import pandas as pd

# Monter Google Drive
drive.mount('/content/drive')

# Chemins des fichiers
base_path = "/content/drive/MyDrive/"

files = {
    "globalai": "globalai index.csv",
    "toolai": "toolAI.csv",
    "impactai": "impactAI.csv",
    "pib": "PIB.csv"
}

# Chargement des datasets
df_globalai = pd.read_csv(base_path + files["globalai"])
df_toolai = pd.read_csv(base_path + files["toolai"])
df_impactai = pd.read_csv(base_path + files["impactai"])
df_pib = pd.read_csv(base_path + files["pib"], header=2)

# Affichage des aperçus
print("Aperçu df_globalai:")
display(df_globalai.head())
print("\nAperçu df_toolai:")
display(df_toolai.head())
print("\nAperçu df_impactai:")
display(df_impactai.head())
print("\nAperçu df_pib:")
display(df_pib.head())
```

```python
# =====================================================================
# ÉTAPE 2  : NETTOYAGE, AGRÉGATION ET HARMONISATION DES TEXTES
# =====================================================================

# 1. df_globalai : pays + score total AI
df_globalai_reduit = df_globalai[['Country', 'Total score']].rename(
    columns={'Total score': 'Global AI Score', 'Country': 'country'}
)

# 2. df_pib : pays + PIB 2023
df_pib_reduit = df_pib[['Country Name', '2023']].rename(
    columns={'2023': 'GDP', 'Country Name': 'country'}
)

# 3. df_toolai : REGRAUPEMENT PAR PAYS ET MOYENNE (Évite les lignes doublées pour un même pays)
df_toolai_reduit = df_toolai.groupby('country')['adoption_rate'].mean().reset_index().rename(
    columns={'adoption_rate': 'Tool Adoption Rate'}
)

# 4. df_impactai : Regroupement par pays et moyenne de productivité
df_impactai_agg = df_impactai.groupby('Country')['Productivity Change (%)'].mean().reset_index().rename(
    columns={'Country': 'country', 'Productivity Change (%)': 'Avg Productivity Change (%)'}
)

# 5. Standardisation invisible : on force tout en minuscules et on nettoie les espaces
for df in [df_globalai_reduit, df_pib_reduit, df_toolai_reduit, df_impactai_agg]:
    df['country'] = df['country'].astype(str).str.strip().str.lower()

print("✅ Étape 2 corrigée avec succès ! Les données d'adoption sont maintenant agrégées par pays.")
display(df_toolai_reduit.head())
```

```python
# ==========================================
# ÉTAPE 3 : FUSION DES DATASETS (LEFT JOIN)
# ==========================================

# On fusionne tout en se basant sur les pays de l'index Global AI
df_merge1 = df_globalai_reduit.merge(df_pib_reduit, on='country', how='left')
df_merge2 = df_merge1.merge(df_toolai_reduit, on='country', how='left')
df_final = df_merge2.merge(df_impactai_agg, on='country', how='left')

# On remet la première lettre des pays en majuscule pour que ce soit beau sur nos graphiques
df_final['country'] = df_final['country'].str.title()

# Affichage des informations de contrôle
print(f"✅ Nombre de pays retenus pour l'analyse : {df_final['country'].nunique()}")
print("\nAperçu des données fusionnées :")
display(df_final.head())
```

```python
# =====================================================================
# ÉTAPE 4
# =====================================================================
from sklearn.model_selection import train_test_split

# 1. Copie de sécurité de notre dataframe fusionné
df_model = df_final.copy()

# 2. Définition de nos variables explicatives (X) et de notre cible à prédire (y)
features = ['GDP', 'Tool Adoption Rate', 'Avg Productivity Change (%)']
target = 'Global AI Score'

# On sépare les caractéristiques de la cible
X = df_model[features]
y = df_model[target]

# 3. Séparation rigoureuse : 70% pour l'entraînement, 30% pour le test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. IMPUTATION PROPRE : On calcule la moyenne sur le Train et on l'applique partout
for col in features:
    # Calcul de la moyenne uniquement sur le sous-ensemble d'entraînement
    valeur_moyenne_train = X_train[col].mean()

    # Remplacement des NaN dans le Train et dans le Test avec cette valeur
    X_train[col] = X_train[col].fillna(valeur_moyenne_train)
    X_test[col] = X_test[col].fillna(valeur_moyenne_train)

print("✅ Étape 4 complétée avec succès !")
print(f"Nombre de pays pour entraîner le modèle : {X_train.shape[0]}")
print(f"Nombre de pays pour tester le modèle : {X_test.shape[0]}")
```

```python
# =====================================================================
# ÉTAPE 5 : ENTRAÎNEMENT ET COMPARAISON RÉELLE DES MODÈLES
# =====================================================================
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# 1. Initialisation des deux modèles
modele_lineaire = LinearRegression()
modele_foret = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)

# 2. Entraînement sur le jeu de données d'entraînement (Train)
modele_lineaire.fit(X_train, y_train)
modele_foret.fit(X_train, y_train)

# 3. Prédictions sur le jeu de données de test (que les modèles n'ont jamais vu)
pred_lineaire = modele_lineaire.predict(X_test)
pred_foret = modele_foret.predict(X_test)

# 4. Calcul des métriques de performance
performance_df = pd.DataFrame({
    'Modèle': ['Régression Linéaire', 'Random Forest'],
    'RMSE': [
        np.sqrt(mean_squared_error(y_test, pred_lineaire)),
        np.sqrt(mean_squared_error(y_test, pred_foret))
    ],
    'R² (Coefficient de détermination)': [
        r2_score(y_test, pred_lineaire),
        r2_score(y_test, pred_foret)
    ]
})

print("=== Comparaison des Performances Réelles (Sans Tricher) ===")
display(performance_df)
```

```python
# =====================================================================
# ÉTAPE 6 : SEGMENTATION (CLUSTERING K-MEANS) ET PRÉPARATION DU DASHBOARD
# =====================================================================
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1. On prépare une copie globale propre pour notre analyse de groupes
df_clustering = df_final.copy()

# Pour la cartographie et le dashboard, on remplit les NaN par la moyenne générale de chaque colonne
for col in features:
    df_clustering[col] = df_clustering[col].fillna(df_clustering[col].mean())

# 2. Normalisation des données (Standardisation)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clustering[features])

# 3. Application du K-Means pour séparer les pays en 3 profils distincts
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_clustering['Cluster'] = kmeans.fit_predict(X_scaled)

# 4. Affichage du profil moyen de chaque cluster pour comprendre nos groupes
print("=== Profil Moyen des 3 Groupes (Clusters) de Pays ===")
profils = df_clustering.groupby('Cluster')[features + ['Global AI Score']].mean()
display(profils)

# 5. SAUVEGARDE DU FICHIER
df_clustering.to_csv("ia_adoption_clusters.csv", index=False)
print("\n✅ Fichier 'ia_adoption_clusters.csv' ")
```
